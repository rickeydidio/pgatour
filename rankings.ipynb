```python
import pandas as pd
rankings = pd.read_excel('/Users/rickeydidio/Desktop/Development/PGA/rankings.xlsx', header=0, sheet_name='Money')
rankings.head()
```

```python
clean = rankings.dropna(thresh=3)
print(clean)
```

```python
import numpy as np
realclean = clean[clean.MoneyRank < 50]
```

```python
import seaborn as sns
correlation = sns.regplot(x='MoneyRank', y='SGTee', data=realclean)
correlation.set(xlabel='Money List Ranking',
                ylabel='Strokes Gained Off the Tee',
                title='Money List Ranking vs SG:OTT with Linear Regression Model Fit')
```

```python
correlation = sns.regplot(x='MoneyRank', y='SGPutting', data=realclean)
correlation.set(xlabel='Money List Ranking',
                ylabel='Strokes Gained Putting',
                title='Money List Ranking vs SG:Putting with Linear Regression Model Fit')
```

```python
correlation = sns.regplot(x='MoneyRank', y='SGGreen', data=realclean)
correlation.set(xlabel='Money List Ranking',
                ylabel='Strokes Gained Around the Green',
                title='Money List Ranking vs SG:Green with Linear Regression Model Fit')
```

```python
correlation = sns.regplot(x='SGTee', y='SGPutting', data=realclean)
correlation.set(xlabel='Strokes Gained Off the Tee',
                ylabel='Strokes Gained Putting',
                title='SG:OTT vs SG:Putting with Linear Regression Model Fit')
```
